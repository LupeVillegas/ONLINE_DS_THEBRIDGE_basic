### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)
2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
3. El scatter de área de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

## Ciudades más pobladas

```python
#Construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro.
#Las ciudades más pobladas, con los numeros de las poblaciones 
#(tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento ha a "center"). 
#(Colores: azul, gris, aceituna, marrón, con trasparencia)

ciudades_pobladas = df.groupby('city')['population_total'].sum().nlargest(5).reset_index()
ciudad = ciudades_pobladas['city'].to_list()
poblacion = ciudades_pobladas['population_total'].to_list()
poblacion = list(map(lambda x: x / (1000), poblacion))

fig, ax = plt.subplots()

colores = ['blue','gray', 'tomato','olive', 'brown']
ax.bar(ciudad, poblacion, color = colores, alpha = 0.6)
for x in range(len(ciudad)): #sacamos los indices de la lista que corresponden al orden de la barras
    ax.text(ciudad[x], poblacion[x] + 100 , str(poblacion[x]), ha= 'center')
    
ax.set_ylim(0, 5000)
ax.set_title('Ciudades más pobladas de California')
ax.set_ylabel('Habitantes (miles)')
ax.set_xlabel('Ciudades');
```

 ## Histograma  del área total ocupada por las ciudades

```python
#construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro.
# Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
fig, ax = plt.subplots(figsize = (6,6))
area = df['area_total_km2'].to_list()

ax.hist(area, bins = 50, color = 'steelblue', alpha = 0.5, label = 'Hist Area')
formato = {
    'family' : 'Bradley Hand itc',
    'size' : 18
}
ax.set_ylabel('Número de ciudades', fontdict = formato) 
ax.set_xlabel('Superfície ($km^2$)', fontdict = formato)
ax.legend(loc='upper right');
```

## Área de agua de un ciudad con su poblacion

```python
#construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro.
#El scatter de área de agua de un ciudad con su poblacion
fig, ax = plt.subplots(figsize = (6,5))
area = df['area_water_km2']
poblacion = df['population_total']
poblacion = list(map(lambda x: x / (1000), poblacion))

ax.scatter(area, poblacion, color = 'green', alpha = 0.4)
props_flecha = {
    'arrowstyle' : '->',
    'connectionstyle' : 'arc3' 
}
ax.annotate('Outliers', xy = (87,3855), xytext = (60,3500), arrowprops= props_flecha)
ax.annotate('Outliers', xy = (122, 1360), xytext = (60,3500), arrowprops= props_flecha)
ax.set_ylabel('Población (miles habitantes)')
ax.set_xlabel('Área de agua en la ciudad ($km^2$)')
ax.set_title('Población por kilometro cuadrado de agua');
```

## Boxplot de la altura de las ciudades

```python
#construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro.
#El boxplot de la altura de las ciudades
fig, ax = plt.subplots()
ax.boxplot(df['elevation_m'], vert = False,patch_artist=True, boxprops=dict(facecolor = 'r'), medianprops = dict(color = 'k'))

ax.set_xlabel('Altitud (m)')
ax.set_title('Diagrama de caja de la Altitud de las ciudades de California');
```

## Gráficos

```python
fig, ax = plt.subplots(2,2, figsize = (20,20))
fig.subplots_adjust(hspace = 0.2 , wspace= 0.2)

ciudad = ciudades_pobladas['city'].to_list()
poblacion = ciudades_pobladas['population_total'].to_list()
poblacion = list(map(lambda x: x / (1000), poblacion))
colores = ['blue','gray', 'tomato','olive', 'brown']
ax[0,0].bar(ciudad, poblacion, color = colores, alpha = 0.6)
for x in range(len(ciudad)): #sacamos los indices de la lista que corresponden al orden de la barras
    ax[0,0].text(ciudad[x], poblacion[x] + 100 , str(poblacion[x]), ha= 'center')
ax[0,0].set_ylim(0, 5000)
ax[0,0].set_title('Ciudades más pobladas de California')
ax[0,0].set_ylabel('Habitantes (miles)')
ax[0,0].set_xlabel('Ciudades');

#===================
area = df['area_total_km2'].to_list()
ax[0,1].hist(area, bins = 50, color = 'steelblue', alpha = 0.5, label = 'Hist Area')
formato = {
    'family' : 'Bradley Hand itc',
    'size' : 18
}
ax[0,1].set_ylabel('Número de ciudades', fontdict = formato) 
ax[0,1].set_xlabel('Superfície ($km^2$)', fontdict = formato)
ax[0,1].legend(loc='upper right');


#===================
area = df['area_water_km2']
poblacion = df['population_total']
poblacion = list(map(lambda x: x / (1000), poblacion))

ax[1,0].scatter(area, poblacion, color = 'green', alpha = 0.4)
props_flecha = {
    'arrowstyle' : '->',
    'connectionstyle' : 'arc3' 
}
ax[1,0].annotate('Outliers', xy = (87,3855), xytext = (60,3500), arrowprops= props_flecha)
ax[1,0].annotate('Outliers', xy = (122, 1360), xytext = (60,3500), arrowprops= props_flecha)
ax[1,0].set_ylabel('Población (miles habitantes)')
ax[1,0].set_xlabel('Área de agua en la ciudad ($km^2$)')
ax[1,0].set_title('Población por kilometro cuadrado de agua');

#===================
ax[1,1].boxplot(df['elevation_m'], vert = False,patch_artist=True, boxprops=dict(facecolor = 'r'), medianprops = dict(color = 'k'))
ax[1,1].set_xlabel('Altitud (m)')
ax[1,1].set_title('Diagrama de caja de la Altitud de las ciudades de California');
plt.show()
```
