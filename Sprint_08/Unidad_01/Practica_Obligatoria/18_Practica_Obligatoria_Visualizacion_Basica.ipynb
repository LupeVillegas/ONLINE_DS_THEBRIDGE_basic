### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:

## Ciudades ms pobladas

```python
#Construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro.
#Las ciudades ms pobladas, con los numeros de las poblaciones 
#(tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento ha a "center"). 
#(Colores: azul, gris, aceituna, marrn, con trasparencia)

ciudades_pobladas = df.groupby('city')['population_total'].sum().nlargest(5).reset_index()
ciudad = ciudades_pobladas['city'].to_list()
poblacion = ciudades_pobladas['population_total'].to_list()
poblacion = list(map(lambda x: x / (1000), poblacion))

fig, ax = plt.subplots()

colores = ['blue','gray', 'tomato','olive', 'brown']
ax.bar(ciudad, poblacion, color = colores, alpha = 0.6)
for x in range(len(ciudad)): #sacamos los indices de la lista que corresponden al orden de la barras
    ax.text(ciudad[x], poblacion[x] + 100 , str(poblacion[x]), ha= 'center')
    
ax.set_ylim(0, 5000)
ax.set_title('Ciudades ms pobladas de California')
ax.set_ylabel('Habitantes (miles)')
ax.set_xlabel('Ciudades');
```

 ## Histograma  del rea total ocupada por las ciudades

```python
#construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro.
# Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
fig, ax = plt.subplots(figsize = (6,6))
area = df['area_total_km2'].to_list()

ax.hist(area, bins = 50, color = 'steelblue', alpha = 0.5, label = 'Hist Area')
formato = {
    'family' : 'Bradley Hand itc',
    'size' : 18
}
ax.set_ylabel('Nmero de ciudades', fontdict = formato) 
ax.set_xlabel('Superfcie ($km^2$)', fontdict = formato)
ax.legend(loc='upper right');
```

## rea de agua de un ciudad con su poblacion

```python
#construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro.
#El scatter de rea de agua de un ciudad con su poblacion
fig, ax = plt.subplots(figsize = (6,5))
area = df['area_water_km2']
poblacion = df['population_total']
poblacion = list(map(lambda x: x / (1000), poblacion))

ax.scatter(area, poblacion, color = 'green', alpha = 0.4)
props_flecha = {
    'arrowstyle' : '->',
    'connectionstyle' : 'arc3' 
}
ax.annotate('Outliers', xy = (87,3855), xytext = (60,3500), arrowprops= props_flecha)
ax.annotate('Outliers', xy = (122, 1360), xytext = (60,3500), arrowprops= props_flecha)
ax.set_ylabel('Poblacin (miles habitantes)')
ax.set_xlabel('rea de agua en la ciudad ($km^2$)')
ax.set_title('Poblacin por kilometro cuadrado de agua');
```

## Boxplot de la altura de las ciudades

```python
#construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro.
#El boxplot de la altura de las ciudades
fig, ax = plt.subplots()
ax.boxplot(df['elevation_m'], vert = False,patch_artist=True, boxprops=dict(facecolor = 'r'), medianprops = dict(color = 'k'))

ax.set_xlabel('Altitud (m)')
ax.set_title('Diagrama de caja de la Altitud de las ciudades de California');
```
