### #1.2

Deshazte de las columnas "deck", "survived", "pclass" y "embarked" porque no las emplearemos, la primera por su cantidad de nulos y el resto porque son variantes de otras que ya existen y cuyos valores resultan ms expresivos a la hora de hacer un anlisis.

### #1.3

**Relacin aircompany-duracion**

```python
plot_grouped_histograms(df_air_jun, "aircompany","duracion",group_size= 2)
```

```python
grupos = df_air_jun['aircompany'].unique() 
ingresos_por_company = [df_air_jun[df_air_jun['aircompany'] == grupo]['duracion'] for grupo in grupos] 
f, p = stats.f_oneway(*ingresos_por_company)  
print("F:", f, "P-Value:", '%.2E' %Decimal(p))
```

**Anlisis de resultados**. Los histogramas y diagramas de caja mostraban que la compaa TabarAir tiene el rango de ingresos ms bajo, sin embargo, es la compaa con ms vuelos. 
Por lo tanto, haba que explorar la causa de esta aparente discrepancia. Se analizaron las relaciones entre aerolnea y diferentes variables, concluyendo que existe una relacin estadstica significativa (p<0.05) entre aerolnea-destino, aerolnea-escala, aerolnea-ingresos, aerolnea-distancia y aerolnea-duracion. Esto se interpreta en el sentido de que la empresa depende tanto del destino como de la escala, los ingresos, la distancia y la duracin. No todas las compaas viajan a todos los destinos, hacen escalas o recorren la misma distancia.
Si analizamos el caso concreto de TabarAir, hay una mayor frecuencia de vuelos sin escalas, la mayora de ellos son vuelos de corta distancia (<2500 km) que claramente tardan menos tiempo. As que, aunque hacen ms vuelos, sus vuelos ms cortos son probablemente ms baratos, por lo que obtienen menos ingresos.
