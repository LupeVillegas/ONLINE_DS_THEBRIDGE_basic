### #1.2

Deshazte de las columnas "deck", "survived", "pclass" y "embarked" porque no las emplearemos, la primera por su cantidad de nulos y el resto porque son variantes de otras que ya existen y cuyos valores resultan más expresivos a la hora de hacer un análisis. 

```python
eliminar_col = ["deck", "survived", "pclass", "embarked"]
for col in eliminar_col:
    df_titanic = df_titanic.drop(col, axis=1)
```

```python
df_titanic.columns
```

### #1.3

Suma las columnas "parch" y "sibsp" para tener el número de parientes de cada pasajero, guarda el resultado en otra columna "family_members", luego deshazte de "parch" y "sibsp"

```python
df_titanic["family_members"] = df_titanic["parch"] + df_titanic["sibsp"]
df_titanic= df_titanic.drop(["parch", "sibsp"], axis=1)
```

```python
df_titanic.head(5)
```

### #1.4

Imputa la moda a los valores nulos de "embark_town"

```python
df_titanic.loc[df_titanic["embark_town"].isna(), "embark_town"] = df_titanic["embark_town"].mode()[0] 
```

### #1.5

Imputa la media a los valores faltantes de "age", si quieres puedes hacer algo más preciso (por ejemplo considerando además la columna "who")

```python
df_titanic.loc[df_titanic["age"].isna(), "age"] = df_titanic["age"].mean() 
```

```python
grupos = df_air_jun['aircompany'].unique() 
ingresos_por_company = [df_air_jun[df_air_jun['aircompany'] == grupo]['ingresos'] for grupo in grupos] 
f, p = stats.f_oneway(*ingresos_por_company)  
print("F:", f, "P-Value:", '%.2E' %Decimal(p))
```

**Relación aircompany-distancia**

```python
plot_grouped_histograms(df_air_jun, "aircompany","distancia",group_size= 2)
```

**Relación aircompany-duracion**

```python
plot_grouped_histograms(df_air_jun, "aircompany","duracion",group_size= 2)
```

```python
grupos = df_air_jun['aircompany'].unique() 
ingresos_por_company = [df_air_jun[df_air_jun['aircompany'] == grupo]['duracion'] for grupo in grupos] 
f, p = stats.f_oneway(*ingresos_por_company)  
print("F:", f, "P-Value:", '%.2E' %Decimal(p))
```

**Análisis de resultados**. Los histogramas y diagramas de caja mostraban que la compañía TabarAir tiene el rango de ingresos más bajo, sin embargo, es la compañía con más vuelos. 
Por lo tanto, había que explorar la causa de esta aparente discrepancia. Se analizaron las relaciones entre aerolínea y diferentes variables, concluyendo que existe una relación estadística significativa (p<0.05) entre aerolínea-destino, aerolínea-escala, aerolínea-ingresos, aerolínea-distancia y aerolínea-duracion. Esto se interpreta en el sentido de que la empresa depende tanto del destino como de la escala, los ingresos, la distancia y la duración. No todas las compañías viajan a todos los destinos, hacen escalas o recorren la misma distancia.
Si analizamos el caso concreto de TabarAir, hay una mayor frecuencia de vuelos sin escalas, la mayoría de ellos son vuelos de corta distancia (<2500 km) que claramente tardan menos tiempo. Así que, aunque hacen más vuelos, sus vuelos más cortos son probablemente más baratos, por lo que obtienen menos ingresos.
